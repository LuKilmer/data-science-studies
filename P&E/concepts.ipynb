```python
#imports
import statistics
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
#dados dos alunos
altura = [1.65, 1.70, 1.75, 1.68, 1.80, 1.62, 1.73, 1.67, 1.69, 1.71,
          1.78, 1.76, 1.58, 1.72, 1.79, 1.66, 1.74, 1.81, 1.64, 1.77,
          1.60, 1.83, 1.68, 1.75, 1.82, 1.70, 1.63, 1.76, 1.69, 1.72]
altura.sort()
peso = [60, 70, 55, 65, 75, 58, 68, 62, 64, 69,
        72, 80, 50, 73, 77, 66, 70, 78, 57, 74,
        53, 85, 61, 72, 79, 68, 55, 74, 62, 70]
peso.sort()
idade = [16, 17, 18, 16, 17, 16, 17, 17, 16, 18,
         17, 16, 17, 17, 17, 18, 16, 17, 16, 18,
         17, 17, 16, 17, 18, 16, 17, 16, 17, 17]
idade.sort()
media = [7.5, 8.0, 6.5, 7.8, 9.0, 6.0, 7.2, 7.6, 7.0, 8.5,
         8.8, 9.5, 6.2, 8.7, 8.9, 7.4, 8.0, 9.2, 6.8, 8.3,
         6.5, 9.3, 7.0, 8.2, 9.1, 8.0, 6.8, 8.4, 7.2, 8.1]
media.sort()
sexo = ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F',
        'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F',
        'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F']
```

```python
contagem_sexo = Counter(sexo)
contagem_idade = Counter(idade)
print("Distribuição de Sexo:", contagem_sexo)
print("Distribuição de Idade:", contagem_idade)
media_altura = statistics.mean(altura)
media_peso = statistics.mean(peso)
media_idade = statistics.mean(idade)
media_media = statistics.mean(media)
print(f"Média altura: {round(media_altura,2)}")
print(f"Média peso: {round(media_peso,2)}")
print(f"Média idade: {round(media_idade,2)}")
print(f"Média media: {round(media_media,2)}")
desvio_padrao_altura = statistics.stdev(altura)
desvio_padrao_peso = statistics.stdev(peso)
desvio_padrao_idade = statistics.stdev(idade)
desvio_padrao_media = statistics.stdev(media)
print("Desvio Padrão de Altura:", desvio_padrao_altura)
print("Desvio Padrão de Peso:", desvio_padrao_peso)
print("Desvio Padrão de Idade:", desvio_padrao_idade)
print("Desvio Padrão de Média:", desvio_padrao_media)
variancia_altura = statistics.variance(altura)
variancia_peso = statistics.variance(peso)
variancia_idade = statistics.variance(idade)
variancia_media = statistics.variance(media)
print(f"Variância altura: {round(variancia_altura,4)}")
print(f"Variância peso: {round(variancia_peso,4)}")
print(f"Variância idade: {round(variancia_idade,4)}")
print(f"Variância media: {round(variancia_media,4)}")
```

```python
import matplotlib.pyplot as plt
plt.hist(peso, bins=10, edgecolor='black',color='yellow')  
plt.xlabel('Peso')
plt.ylabel('Frequência')
plt.title('Distribuição de Pesos')
plt.show()
plt.hist(altura, bins=10, edgecolor='black',color='blue')  
plt.xlabel('Altura')
plt.ylabel('Frequência')
plt.title('Distribuição de Altura')
plt.show()
labels = list(contagem_idade.keys())
sizes = list(contagem_idade.values())
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
num_bins = 10
fig, ax = plt.subplots()
n, bins, patches = ax.hist(peso, num_bins, density=True, edgecolor='black',color='yellow')
y = ((1 / (np.sqrt(2 * np.pi) * desvio_padrao_peso)) *
     np.exp(-0.5 * ((bins - media_peso) / desvio_padrao_peso)**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Peso')
ax.set_ylabel('Densidade de Probabilidade')
ax.set_title(f'Histograma da variável "peso": $\mu={media_peso:.2f}$, $\sigma={desvio_padrao_peso:.2f}$')
fig.tight_layout()
plt.show()
num_bins = 10
fig, ax = plt.subplots()
n, bins, patches = ax.hist(altura, num_bins, density=True, edgecolor='black',color='blue')
y = ((1 / (np.sqrt(2 * np.pi) * desvio_padrao_altura)) *
     np.exp(-0.5 * ((bins - media_altura) / desvio_padrao_altura)**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Peso')
ax.set_ylabel('Densidade de Probabilidade')
ax.set_title(f'Histograma da variável "altura": $\mu={media_altura:.2f}$, $\sigma={desvio_padrao_altura:.2f}$')
fig.tight_layout()
plt.show()
```

```python
import numpy as np

idade = np.array([2, 4, 6, 8, 10])
pulso = np.array([112, 104, 100, 92, 88])

correlacao = np.corrcoef(idade, pulso)[0, 1]
print(correlacao)
n = len(idade)
sum_x = np.sum(idade)
sum_y = np.sum(pulso)
sum_xy = np.sum(idade * pulso)
sum_x_squared = np.sum(idade**2)

# Calculando os coeficientes da reta de regressão
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
b = (sum_y - m * sum_x) / n

print("Coeficiente Angular (m):", m)
print("Coeficiente Linear (b):", b)
print(b+m*9)
```
